## test image

```python
import numpy as np
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt
```

```python
img = Image.new('RGB', (200, 200), (255, 0, 0))
bboxes = np.array([[20, 20, 60, 60, 0, 0], [30, 30, 100, 100, 0, 0]], dtype=np.float32)
points = np.array([[50, 50, 0, 0], [150, 150, 0, 0]], dtype=np.float32)

draw = ImageDraw.Draw(img)
for i in range(bboxes.shape[0]):
    draw.rectangle(bboxes[i, 0:4])
for i in range(points.shape[0]):
    draw.point(points[i, 0:2])
plt.imshow(np.array(img));
```

```python
img_, bboxes_, points_ = image.resize_image(img, (1024, 768), bboxes=bboxes, points=points)
draw = ImageDraw.Draw(img)
for i in range(bboxes_.shape[0]):
    draw.rectangle(bboxes_[i, 0:4])
for i in range(points_.shape[0]):
    draw.point(points_[i, 0:2])
plt.imshow(np.array(img_));
```
